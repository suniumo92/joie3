```python
#去除json顶层嵌套，多层级list呈现
f1 = urllib.request.urlopen("http://data.cstr.ed.ac.uk/globalvox/asr/2017/20170101_014100_global_news_podcast.asr.json")
objects1 = ijson.items(f1, "segments.item")
segments1 = list (objects1)
```

```python
f2 = urllib.request.urlopen("http://data.cstr.ed.ac.uk/globalvox/asr/2016/20160101_105100_best_of_today.asr.json")
objects2 = ijson.items(f2, "segments.item")
segments2 = list (objects2)
```

```python
f3 = urllib.request.urlopen("http://data.cstr.ed.ac.uk/globalvox/asr/2016/20160101_130000_the_fifth_floor.asr.json")
objects3 = ijson.items(f3, "segments.item")
segments3 = list (objects3)
```

```python
#解析被复合嵌套的数据，进一步减少层级嵌套，list呈现
spokenWords1 = [words1["spokenWords"] for words1 in segments1]
spokenWords2 = [words2["spokenWords"] for words2 in segments2]
spokenWords3 = [words3["spokenWords"] for words3 in segments3]
```

```python
spokenWords = spokenWords1 + spokenWords2 + spokenWords3
```

```python
import itertools      # "压平"多重嵌套的list，得到可整合的一个list格式：[{a: 1, b: 2}, {a: 3, b: 4}]
spokenWords = list(itertools.chain.from_iterable(spokenWords))
```

```python
spokenWords
```

```python
#整合数据，table呈现
cols = list(spokenWords[0].keys())
rows = []
indices = []
for index, d in enumerate(spokenWords):
    indices.append(index)
    values = []
    for col in cols:
        values.append(d[col])
    rows.append(values)

df = pd.DataFrame(rows, index = indices, columns = cols)
df
```

```python
wordId = list(df.wordId)
myset = set(wordId)
for item in myset:
    print("the %s has found %s" %(item,wordId.count(item)))
```
